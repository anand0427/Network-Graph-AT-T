```python
import pandas as pd
import numpy as np
```

```python
import plotly
```

```python
import networkx as nx
```

```python
network_df = pd.read_csv("../network_data.csv") 
```

```python
network_df.head()
```

```python
A = list(network_df["source_ip"].unique())
```

```python
B = list(network_df["destination_ip"].unique())
```

```python
node_list = set(A+B)
```

```python
G = nx.Graph(day="IP")
```

```python
for i in node_list:
    G.add_node(i)
```

```python
G.nodes()
```

```python
for i,j in network_df.iterrows():
    G.add_edges_from([(j["source_ip"],j["destination_ip"])])
```

```python
G.edges()
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
import plotly.graph_objs as go
```

```python
init_notebook_mode(connected=True)
```

```python
pos = nx.spring_layout(G, k=0.5, iterations=50)
```

```python
pos
```

```python
for n, p in pos.items():
    G.node[n]['pos'] = p
```

```python
edge_trace = go.Scatter(
    x=[],
    y=[],
    line=dict(width=0.5,color='#888'),
    hoverinfo='none',
    mode='lines')

for edge in G.edges():
    x0, y0 = G.node[edge[0]]['pos']
    x1, y1 = G.node[edge[1]]['pos']
    edge_trace['x'] += tuple([x0, x1, None])
    edge_trace['y'] += tuple([y0, y1, None])
```

```python
node_trace = go.Scatter(
    x=[],
    y=[],
    text=[],
    mode='markers',
    hoverinfo='text',
    marker=dict(
        showscale=True,
        colorscale='RdBu',
        reversescale=True,
        color=[],
        size=15,
        colorbar=dict(
            thickness=10,
            title='Node Connections',
            xanchor='left',
            titleside='right'
        ),
        line=dict(width=0)))

for node in G.nodes():
    x, y = G.node[node]['pos']
    node_trace['x'] += tuple([x])
    node_trace['y'] += tuple([y])
```

```python
for node, adjacencies in enumerate(G.adjacency()):
    node_trace['marker']['color']+=tuple([len(adjacencies[1])])
    node_info = adjacencies[0] +' # of connections: '+str(len(adjacencies[1]))
    node_trace['text']+=tuple([node_info])
```

```python

fig = go.Figure(data=[edge_trace, node_trace],
             layout=go.Layout(
                title='<br>AT&T network connections',
                titlefont=dict(size=16),
                showlegend=False,
                hovermode='closest',
                margin=dict(b=20,l=5,r=5,t=40),
                annotations=[ dict(
                    text="No. of connections",
                    showarrow=False,
                    xref="paper", yref="paper") ],
                xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                yaxis=dict(showgrid=False, zeroline=False, showticklabels=False)))

iplot(fig, filename='networkx')
```
